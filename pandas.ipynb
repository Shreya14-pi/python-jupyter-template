```python
print('Count of the Profit column:', Orders['Profit'].count())
print("Minimum of the Profit column:",Orders['Profit'].min())
print("Maximum of the Profit column:", Orders['Profit'].max())
print("Average/Mean of the Profit column:", Orders['Profit'].mean())
print('Standard Deviation of the Profit column:', Orders['Profit'].std())
#25% of the total Profit data count are below Quartile 1 value of 1.76
print('Quartile 1 value of the Profit column:', Orders['Profit'].quantile(0.25))
#50% of the total Profit data count are below Quartile 2/Median value of 8.69
print("Quartile 2/Median value of the Profit column:", Orders['Profit'].quantile(0.50))
#75% of the total Profit data count are below Quartile 3 value of 29.29
print('Quartile 3 value of the Profit column:', Orders['Profit'].quantile(0.75))
```
